```python
data.drop(['Gender', 'Education Level', 'Job Title', 'Age'], axis=1, inplace=True)
```

```python
x = data.drop("Salary", axis=1)
y = data["Salary"]
```

```python
plt.scatter(x,y)
plt.title("Salary base on Years of Experience")
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
y_predict = model.predict(x)
y_predict = model.intercept_ + model.coef_ * x
```

```python
plt.scatter(x,y)
plt.plot(x,y_predict ,color = 'red')
plt.title("Salary base on Years of Experience")
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
r_squared = model.score(x,y)
print(r_squared)
```
